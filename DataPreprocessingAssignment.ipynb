```python
from cmapPy.pandasGEXpress.parse_gct import parse
import matplotlib.pyplot as plt
import pandas as pd
```

# Part 1

```python
def process_gct_file(gct_file_path, summary=False):
    '''
    Reads the inputted GenePattern gct file into a Pandas DataFrame. 
    
    Args:
        gct_file_path: the path to the GenePattern gct file to be processed, as a string
        summary: whether or not to display summary information, as a boolean (defaults to False)
    Returns:
        A Pandas DataFrame containing data from the gct file.
    '''
    df = parse(gct_file_path).data_df
    if summary:
        rows, cols = df.shape
        print(f'Number of rows: {rows}\nNumber of columns: {cols}')
    return df
    
```

# Part 3

```python
def generate_histograms(columns, df):
    '''
    Generates histograms for specified columns in a specified Pandas
    DataFrame.
    
    Args:
        columns: the columns to generate histograms for, as a list of strings
        df: the Pandas DataFrame to pull data from
    Returns:
        None
    '''
    for column in columns:
        plt.figure(figsize=(10, 5))
        plt.hist(df[column], bins=20)
        plt.title(f'Histogram for {column}')
        plt.show()
```

```python
columns_of_interest = ['A7-A0DB-normal', 'A7-A13E-normal', 'BH-A0B3-primary', 'BH-A0B5-primary']
generate_histograms(columns_of_interest, large_df)
```

# Part 4

```python
def generate_summary_df(df):
    '''
    Generates a summary Pandas DataFrame containing sample name, mean, median, 
    and standard deviation for each column in the inputted DataFrame.
    
    Args:
        df: the inputted Pandas DataFrame
    Returns:
        The column summaries, as a Pandas DataFrame where each row is a sample
    
    '''
    summary_df = pd.DataFrame(columns=['Sample Name', 'Mean', 'Median', 'Standard Deviation'])
    
    for column in df.columns:
        sample = df[column]
        mean = sample.mean()
        median = sample.median()
        std_dev = sample.std()
        
        summary_df = summary_df.append({
            'Sample Name': column,
            'Mean': mean,
            'Median': median,
            'Standard Deviation': std_dev
        }, ignore_index=True)
    
    return summary_df
```

```python
generate_summary_df(large_df)
```
